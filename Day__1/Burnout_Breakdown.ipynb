```python
Q1=df.WorkHoursPerWeek.quantile(0.25)
Q3=df.WorkHoursPerWeek.quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df.columns = df.columns.str.strip()
x = df['StressLevel']
y = df['SleepHours']
colors = df['StressLevel']
plt.scatter(x, y, c=colors, cmap='coolwarm', s=20, alpha=0.7)
plt.xlabel('Stress Level')
plt.ylabel('Sleep Hours')
plt.title('Stress Level vs Sleep Hours')
plt.grid(True)
plt.show()
```
